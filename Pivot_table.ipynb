### Pandas Pivot Table
The levels in the pivot table will be stored in multiIndex object on the index and columns of the result DataFrame.

Syntax : DataFrame.pivot_table()

Parameters: values,fill_value,index,margins,columns,dropna,aggfunc

```python
import pandas as pd
```

```python
df=pd.read_csv('Iris.csv')
df.head()
```

```python
df.pivot_table(index='SepalLengthCm').head()
```

```python
df.pivot_table(index='SepalLengthCm',columns='Id').head()
```

```python
df.pivot_table(index='SepalLengthCm',columns='Id',aggfunc='count').head()
```

```python
df.pivot_table(index='SepalLengthCm',columns='Id',aggfunc='max').head()
```

```python
df.pivot_table(index='SepalLengthCm',columns='Id',fill_value=0).head()
```

```python
df.pivot_table(index='SepalLengthCm',columns='Id',dropna=False).head()
```

```python
df.pivot_table(index='SepalLengthCm',columns='Id',margins=True).head()
```

```python
df.pivot_table(index='SepalLengthCm',columns='Id',fill_value=0,aggfunc='sum',margins=True).head()
```
