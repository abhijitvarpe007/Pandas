### Pandas groupby Function
Pandas groupby function is used to split the data into groups based on some criteria.

Operation: Spliting the object, Applying a Function, Combining the result

```python
df.head()
```

```python
gr1=df.groupby(by='Species')
gr1.groups
```

```python
gr2=df.groupby(['Species','SepalLengthCm'])
gr2.groups
```

```python
for Species,df_1 in gr1:
    print(Species)
    print(df_1)
```

```python
list(gr1)
```

```python
dict(list(gr1))
```

```python
df.groupby('SepalLengthCm').get_group(5)
```

```python
df.groupby('SepalLengthCm').get_group(4.9)
```
