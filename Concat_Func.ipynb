### Pandas Concat() Function
Pandas provides various facilities for easily combining togetherSeries,DataFrame and Panel Object


 Syntax: pd.concate()
 
 parameter: objs, axis, join,join_axes,ignore_index,key,sort

```python
import pandas as pd
```

```python
sr1=pd.Series([0,1,2])
sr1
```

```python
sr2=pd.Series([3,4,5])
sr2
```

```python
pd.concat([sr1,sr2])
```

```python
sr3=pd.Series([6,7])
sr3
```

```python
pd.concat([sr1,sr2])
```

```python
df2=pd.DataFrame({'ID':[1,2,3,4],
              "Name":['A','B','C','D'],
             "Class":[5,6,7,8]})
df2
```

```python
df3=pd.DataFrame({'ID':[5,6,7,8],
              "Name":['E','F','G','H'],
             "Class":[9,10,11,12]})
df3
```

```python
pd.concat([df2,df3])
```

```python
pd.concat([df3,df2])
```

```python
pd.concat([df2,df3],axis=1)
```

```python
pd.concat([df2,df3],axis=0,ignore_index=True)
```

```python
df4=pd.DataFrame({'ID':[1,2,3,4],
              "Name":['A','B','C','D'],
             "Class":[5,6,7,8]})
df4
```

```python
df5=pd.DataFrame({'ID':[3,4],
              "Name":['C','D'],
             "Class":[7,8]})
df5
```

```python
pd.concat([df4,df5],axis=1)
```

```python
pd.concat([df4,df5],join='inner',axis=1)
```

```python
pd.concat([df4,df5],axis=1,join='outer')
```

```python
df6=pd.DataFrame({'ID':[5,6,7,8],
              "Name":['E','F','G','H'],
             "Class":[9,10,11,12]})
df6
```

```python
pd.concat([df4,df6],keys=['df4','df6'])
```

```python
pd.concat([df4,df6],keys=['First_df','Second_df'])
```

```python
pd.concat([df4,df6],axis=1,keys=['df4','df6'])
```

```python
df7=pd.DataFrame({'ID':[1,2,3,4],
              "Name":['A','B','C','D'],
             "Class":[5,6,7,8]})
df7
```

```python
df8=pd.DataFrame({"Marks":[40,36,90,67]})
df8
```

```python
pd.concat([df7,df8],sort=False)
```
