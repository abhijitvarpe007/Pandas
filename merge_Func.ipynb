### Pandas merge Function
Pandas merge connects columns or indexes in DataFrame based on one or more keys

```python
import pandas as pd
```

```python
df=pd.DataFrame({'Id':[1,2,3,4],'Class':['A','B','C','D']})
df
```

```python
df1=pd.DataFrame({'Id':[1,2,3,4],'Name':['Abhi','Sham','RAm','Ramesh']})
df1
```

```python
pd.merge(df1,df,on='Id')
```

```python
pd.merge(df,df1,on='Id')
```

```python
df2=pd.DataFrame({'Id':[1,2,5,6],'Name':['Abhi','Sham','RAm','Ramesh']})
```

```python
pd.merge(df1,df2,on='Id')
```

```python
pd.merge(df1,df2,on='Id',how='inner')
```

```python
pd.merge(df1,df2,on='Id',how='left')
```

```python
pd.merge(df1,df2,on='Id',how='right')
```

```python
pd.merge(df1,df2,on='Id',how='outer')
```

```python
pd.merge(df1,df2,on='Id',how='outer',indicator=True)
```
