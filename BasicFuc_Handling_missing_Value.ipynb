### Handling missing values in python

### 1.na_values 2.keep_default_na 3.na.filter

```python
df=pd.read_csv('C:\\Users\\HP\\Data Science\\Pandas\\train.csv')
df.head()
```

```python
# using na_valuves function to convert the NaN values
# to convert male string into NaN values
df=pd.read_csv('C:\\Users\\HP\\Data Science\\Pandas\\train.csv',na_values='male')
df.head()
```

```python
# Specific column to remove unnessceery value using dictionary
df=pd.read_csv('C:\\Users\\HP\\Data Science\\Pandas\\train.csv',na_values={"Sex":'female'})
df.head(5)
```

```python
#Keep Default method
df=pd.read_csv('C:\\Users\\HP\\Data Science\\Pandas\\train.csv',keep_default_na=False)
df.head(5)
```

```python
df=pd.read_csv('C:\\Users\\HP\\Data Science\\Pandas\\train.csv',keep_default_na=True)
df.head(5)
```

```python
# Dataset doesnot contain null value to use the filter method to boost your performance
df=pd.read_csv('C:\\Users\\HP\\Data Science\\Pandas\\train.csv', na_filter=False)
df.head(5)
```

### Methods : 1.isnull() 2.notnull()

```python
# isnull method is represent the missing values
df=pd.read_csv('C:\\Users\\HP\\Data Science\\Pandas\\train.csv')
df.isnull().head()
```

```python
df.isnull().sum() # find number of null values
```

```python
# not null method is represesnt the correct values
df.notnull()
```

```python
# total number of not null values
df.notnull().sum()
```
